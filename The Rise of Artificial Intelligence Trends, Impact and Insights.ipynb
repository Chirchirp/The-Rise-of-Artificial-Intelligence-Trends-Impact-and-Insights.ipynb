##  2. Data Exploration & Cleaning  
###  2.1 Load the Dataset  
- Read the dataset into a Pandas DataFrame.  
- Display the first few rows to understand its structure.  

```python
#imprting Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Reading datasets and displaying top 5 rows

df = pd.read_csv(r"B:\Datasets\Rise of AI\Artificial_Intel_Data.csv")
df.head()
```

### 2.2 Check for Missing Values & Data Types  

```python
# Check for Missing Values:

count_nulls = df.isnull().sum().sum()

if count_nulls > 0:
    print(f"Total Null values are {count_nulls}")
else:
    print("No missing values in the dataset")
```

```python
# Check data types
df.info()
```

### 2.3 Data Cleaning of Percentage Columns

```python
# Identify columns with percentage values
percentage_columns = [col for col in df.columns if df[col].astype(str).str.contains('%').any()]

print("Columns containing % sign:", percentage_columns)
```

```python
# Convert percentage columns to numeric
for col in percentage_columns:
    df[col] = df[col].str.rstrip('%').astype(float)

df
```

##  3. AI Market Growth Analysis  
###  3.1 AI Software Revenue and Global AI Market Value Over Time 

```python
# These line plots reveal the growth trends for AI software revenue and the overall global AI market value over the years.

plt.figure(figsize=(12, 5))
sns.lineplot(x='Year', y='AI Software Revenue(in Billions)', data=df, marker='o', label='Software Revenue', color='#1f77b4')
sns.lineplot(x='Year', y='Global AI Market Value(in Billions)', data=df, marker='s', label='Global Market Value', color='#ff7f0e')
plt.title("Growth of AI Software Revenue and Global AI Market Value")
plt.xlabel("Year")
plt.ylabel("Billions USD")
plt.legend(title="Metric", loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

**Explosive Growth in AI Investment:**  
- Both AI software revenue and global market value have shown significant, steady growth. This demonstrates the expanding role of AI in driving business value

## 3.2 Estimated revenue increase from AI

```python
# Estimated Revenue Increase from AI
plt.figure(figsize=(12, 5))
ax1 = sns.barplot(x='Year', y='Estimated Revenue Increase from AI (trillions USD)', data=df, color='#17becf')
plt.title("Estimated Revenue Increase from AI")
plt.xlabel("Year")
plt.ylabel("Revenue Increase (Trillions USD)")
for p in ax1.patches:
    ax1.annotate(f"{p.get_height():.2f}", 
                 (p.get_x() + p.get_width() / 2, p.get_height()), 
                 ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

**Revenue Generated from AI:**  
- There is a steady increase in the revenue being generated by AI industries, indicating a rapid shift in business operations towards AI.

###  3.3 Organizational AI Strategy: Current Use vs. Future Plans

```python
# Bar charts comparing organizations currently using AI versus those planning to implement AI offer insights into market readiness and future growth potential.

plt.figure(figsize=(12, 5))
width = 0.4  # bar width
years = df['Year']
x = np.arange(len(years))

# Plotting side-by-side bars
plt.bar(x - width/2, df['Organizations Using AI'], width=width, label='Using AI', color='#1f77b4')
plt.bar(x + width/2, df['Organizations Planning to Implement AI'], width=width, label='Planning to Implement AI', color='#ff7f0e')

plt.title("Organizations Using AI vs. Planning to Implement AI")
plt.xlabel("Year")
plt.ylabel("Percentage of Organizations")
plt.xticks(ticks=x, labels=years)
plt.legend(title="Category", loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

**Strategic Positioning of Organization on AI Implementation**
    
- The number of organizations planning to implement AI is higher than those that have adopted AI suggesting massive market opportunity and a great shift in business operations

##  4. AI Adoption & Business Impact  
###  4.1 AI Adoption Trends Vs Global Expectations

```python
# This plot compares actual AI adoption percentages against global expectations. A gap between the two can highlight opportunities or areas where expectations exceed current uptake.

plt.figure(figsize=(12, 5))
sns.lineplot(x='Year', y='AI Adoption (%)', data=df, marker='^', label='Current Adoption', color='#2ca02c')
sns.lineplot(x='Year', y='Global Expectation for AI Adoption (%)', data=df, marker='v', label='Expected Adoption', color='#d62728')
plt.title("AI Adoption Trends: Current vs. Global Expectations")
plt.xlabel("Year")
plt.ylabel("Adoption (%)")
plt.legend(title="Adoption Metric", loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

**Bridging the Adoption Gap:**  

  - While current AI adoption rates are increasing, global expectations remain higher. This indicates room for accelerated implementation and further innovation.

##  5. AI’s Impact on the Workforce  
###  5.1 Job Loss vs. Job Creation 

```python
import plotly.graph_objs as go
# Create traces for each line
trace1 = go.Scatter(
    x=df['Year'],
    y=df['Estimated Jobs Eliminated by AI (millions)'],
    mode='lines+markers',
    name='Jobs Eliminated'
)

trace2 = go.Scatter(
    x=df['Year'],
    y=df['Estimated New Jobs Created by AI (millions)'],
    mode='lines+markers',
    name='New Jobs Created'
)

trace3 = go.Scatter(
    x=df['Year'],
    y=df['Estimated Revenue Increase from AI (trillions USD)'],
    mode='lines+markers',
    name='Revenue Increase from AI'
)
trace4 = go.Scatter(
    x=df['Year'],
    y=df['Net Job Loss in the US'],
    mode='lines+markers',
    name='Net Job Loss'
)
# Create a figure and add the traces
fig = go.Figure(data=[trace1, trace2, trace3, trace4])

# Update layout for titles and axes labels
fig.update_layout(
    title='Estimated Jobs Created and Eliminated by AI',
    xaxis_title='Year',
    yaxis_title='Estimated Values (%)',
    legend_title='Metrics',
    template='plotly_white',
    width=1000,
    height=600
)

# Show gridlines
fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)

# Display the plot
fig.show()
```

**Balancing Job Impacts:**  

  - Although AI leads to job elimination in certain areas, it also creates new roles. The net effect—along with targeted upskilling—can be strategically managed.

###  5.2 High-Risk Job Sectors 

```python
# Create traces for each line
trace1 = go.Scatter(
    x=df['Year'],
    y=df['Jobs at High Risk of Automation - Manufacturing'],
    mode='lines+markers',
    name='Manufacturing'
)

trace2 = go.Scatter(
    x=df['Year'],
    y=df['Jobs at High Risk of Automation - Transportation & Storage (%)'],
    mode='lines+markers',
    name='Transportation & Storage'
)

trace3 = go.Scatter(
    x=df['Year'],
    y=df['Jobs at High Risk of Automation - Wholesale & Retail Trade'],
    mode='lines+markers',
    name='Wholesale & Retail Trade',
)
# Create a figure and add the traces
fig = go.Figure(data=[trace1, trace2, trace3])

# Update layout for titles and axes labels
fig.update_layout(
    title='Jobs At High Risks for Ai automation',
    xaxis_title='Year',
    yaxis_title='High Risk Percentage (%)',
    legend_title='Metrics',
    template='plotly_white',
    width=1000,
    height=600
)

# Show gridlines
fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)

# Display the plot
fig.show()
```

**Jobs at high risks due to AI Implementation**

- Organisations should proactively assess the impact of AI on their workforce and develop strategies to manage job displacement.
- Automation risks in transportation, retail, and manufacturing signal areas for workforce retraining and technology integration.

## 7.0 A correlation heatmap

-A correlation heatmap helps identify relationships between variables (e.g., between AI adoption rates and revenue, or between adoption and job impacts). High correlation values could indicate predictive relationships useful for strategic planning.

### **Key observations might include:**
- A strong positive correlation between AI adoption rates and AI-driven revenue growth.
- A correlation between organizations’ strategic prioritization of AI and the overall market value.
- Negative correlations where increased AI adoption might correspond with job losses in certain sectors.

```python
plt.figure(figsize=(14, 10))
corr = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of AI Metrics")
plt.show()
```

**Correlation as a Strategic Tool:**  

  - The strong correlations between AI adoption, revenue growth, and competitive advantages provide data-driven support for investing in AI strategies.
